```python
import pandas as pd 
import requests
import os
import tweepy
import json
import timeit
import json
import numpy as np
```

# Gather

```python

# read in the tweeter file


df=pd.read_csv("twitter-archive-enhanced.csv")
```

```python
df.head(5)
```

```python
# Reading the tsv file programatically

url = "https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv"
response=requests.get(url)
with open(os.path.join(url.split ("/")[-1]), mode='wb') as file:
    file.write(response.content)

```

```python
# Reading the tsv file


idf=pd.read_csv("image-predictions.tsv",'\t')
idf.head()
```

```python

#--- Place holders for the key and tokens 
consumer_key = 'consumer_key'
consumer_secret = 'consumer_secret'
access_token = 'access_token'
access_secret = 'access_secret'


# Gathering data through Tweeter apo

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)

```

```python
# writing the file to a json file
error_tweet_ids=[]

start = timeit.timeit()
with open('tweet_json.txt', 'a' , encoding='utf8') as outfile:
    for id_of_tweet in df.tweet_id:
        try:
            tweet = api.get_status(id_of_tweet, tweet_mode='extended')
            json.dump(tweet._json, outfile)
            outfile.write('\n')
            #print(tweet.id)

        except:
            error_tweet_ids.append(tweet.id)
            print(tweet.id)
            
end = timeit.timeit()
```

```python
# Running the code agian for the tweet ids with error

error_tweet_ids1=[]
with open('tweet_json.txt', 'a' , encoding='utf8') as outfile:
    for id_of_tweet in error_tweet_ids:
        try:
            tweet = api.get_status(id_of_tweet, tweet_mode='extended')
            json.dump(tweet._json, outfile)
            outfile.write('\n')
            print(tweet.id)

        except:
            error_tweet_ids1.append(tweet.id)
```

```python
#with the second run there was no false ids. 
print(error_tweet_ids1)
```

```python

# Reads in each json file and adds it to a list
json_tweets = []
fo=open('tweet_json.txt', 'r')
for line in fo:
    single_tweet=json.loads(line)
    json_tweets.append(single_tweet)
    #jd=pd.read_json(json_tweets, orient='index')
    
fo.close()
```

```python

# read in the data from the json file and and make it into a dataframe
d =[]
for tweet in json_tweets:
    temp = d.append({'id':tweet['id'], 'retweet_count':tweet['retweet_count'],'favorite_count':tweet["favorite_count"] })

jd=pd.DataFrame(d)

# writes the data frame to a csv file for easier reading later
jd.to_csv("json_tweets_dataframe.csv")
```
